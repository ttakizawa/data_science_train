## 密度と等高線図
輪郭や色分けされた領域を使用して３次元データを２次元で表示する方法、Matplotlibでは３つの関数を提供している。<br>
* plt.contour:等高線図
* plt.contourf:等高線図（塗りつぶし）
* plt.imshow:画像を表示

```python
# 必要なパッケージのインポートとjupyter notebookの設定
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

### ３次元関数の可視化
z = f(x,y)の等高線図を作成する。

```python
# 関数fの定義
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

等高線図は、plt.contour関数を使用する。<br>
x値のグリッド、y値のグリッド、z値のグリッドの３つを引数を取る。<br>

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0,5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
print(X.shape)
print(Y.shape)
print(Z.shape)
```

```python
plt.contour(X, Y, Z, colors='black')
```

```python
# 1色だけの場合、デフォルトでは負の値は破線で表示され、正の値は実践で表示される
# 以下ではデータ範囲内を20の等間隔で区切った線で描画するように指定する。
plt.contour(X, Y, Z, 20, cmap='RdGy')
```

```python
# 塗りつぶしに変更
plt.contourf(X, Y, Z, 20, cmap='RdGy')
# カラーバーを追加
plt.colorbar()
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
```
